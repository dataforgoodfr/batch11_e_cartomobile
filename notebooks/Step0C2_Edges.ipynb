```python
from sklearn.neighbors import BallTree

def get_nearest(src_points, candidates, k_neighbors=1):
    """Find nearest neighbors for all source points from a set of candidate points"""
    tree = BallTree(candidates, leaf_size=20, metric='haversine')
    distances, indices = tree.query(src_points, k=k_neighbors)
    if src_points.shape[0] > 1 :
        distances = distances.transpose()
        indices = indices.transpose()
    closest = indices[0]
    closest_dist = distances[0]
    return (closest, closest_dist)
```

```python
import numpy as np


def get_closest_routier( Nodes,nodes_routier,start,end):
    coord = ['x', 'y']
    candidates = nodes_routier[coord]
    closest_routier, _ = get_nearest(
                        np.array([Nodes.loc[start,coord].astype(float),
                                Nodes.loc[end,coord].astype(float)
                                ]),  
                        candidates
    )    
    return nodes_routier.iloc[closest_routier[0]].name, nodes_routier.iloc[closest_routier[1]].name


start = '147124'
end = '146573'

source, target = get_closest_routier(Nodes, Nodes_routier,start,end)
source, target 
```

```python
from shapely import MultiLineString, ops

def get_path_attributes( nodes, graph_routier, start, end):
    source, target = get_closest_routier(nodes, Nodes_routier,start,end)
    path = nx.shortest_path(graph_routier, 
                            source=source, 
                            target=target, 
                            weight='travel_time', 
                            method='dijkstra')
    lines = ox.utils_graph.get_route_edge_attributes(
                    graph_routier, 
                    path, 
                    attribute='geometry', 
                    minimize_key='length', 
                    retrieve_default=None)
    multi_line = MultiLineString(lines)
    merged_line = ops.linemerge(multi_line)

    length = sum(
        ox.utils_graph.get_route_edge_attributes(
                    graph_routier, 
                    path, 
                    attribute='length', 
                    minimize_key='length', 
                    retrieve_default=None)
    )
    
    return merged_line, length


try:
    too_long = joblib.load( 
                os.path.join(CFG.data_path, f'[{CFG.area_name}]too_long_edges_communes.pkl')
            )
except:    
    too_long = []

total = G_communes.number_of_edges()
for n, (u, v, data) in enumerate(G_communes.edges(data=True)):
    # print(u, v,  d)

    state = data.get('length', 0)
    print()
    if (state == 0) and not((u,v) in too_long):
        line, length = get_path_attributes( Nodes, Graph_routier, u, v)
        # print(line, length)
        if length <= 100e3:  # 100 km length limit
            attr = {
                    'geometry': line,
                    'length': length
                }


            G_communes.add_edge(
                    u,
                    v,
                    0,
                    **attr
                )
        else:
            # G_communes.remove_edge(u,v,0)  : Change on edges dict blows edge iterator,
            too_long.append((u,v))
            
        if (n + 1) % 500 == 0:
            joblib.dump(
                ox.graph_to_gdfs(G_communes, nodes=False, edges=True),
                os.path.join(CFG.data_path, f'[{CFG.area_name}]Gedges_communes.pkl')
            )
            joblib.dump(
                too_long, 
                os.path.join(CFG.data_path, f'[{CFG.area_name}]too_long_edges_communes.pkl')
            )
        
    print(f'{n+1}/{total} ~ {(n + 1) / total:.4%} ', end=' '*50 + '\r')

print(f'Too long egdes exclued: {len(too_long)}') 
G_communes.remove_edges_from(too_long)  
```
