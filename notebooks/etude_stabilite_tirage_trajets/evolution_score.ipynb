Ce notebook prend les tirages donnés dans les db_snapshots pour créer un fichier unifié "df_{draw_mode}.csv"

```python
import pandas as pd
```

```python
WorkingFolder = "."
```

```python
draw_mode = "equiprobable_mode" # weighted_mode_2 or equiprobable_mode
```

```python
df_meta = pd.read_csv(f"{WorkingFolder}/meta_{draw_mode}.csv", parse_dates=[3])
```

```python
df_meta
```

```python
s_files = f"db_snapshot_exp_{draw_mode}_" + \
    df_meta.algo_start_date.dt.strftime("%Y-%m-%d_%H_%M_%S") + \
    "_" + \
    df_meta.batch_nb.astype(str)+ "_" + \
    df_meta.update_batch_size.astype(str) 
```

```python
s_files
```

```python
s_files[0]
```

```python
def get_s_data_draw(n_draw):
    s_data = pd.read_csv(
                        f"{WorkingFolder}/{draw_mode}/" + s_files[n_draw], 
                         index_col = 0, 
                         usecols=[1,2],
                         ).squeeze()
    s_data.name = f"draw_{n_draw}"
    return s_data
```

```python
df_data = pd.concat([get_s_data_draw(i) for i in range(len(s_files))],axis=1)
```

```python
df_data
```

```python
df_data.fillna(0).to_csv(f"{WorkingFolder}/df_{draw_mode}.csv")
```
